```python
import pandas as pd
import re
from typing import List, Optional
import logging

def load_sme_csv_to_dataframe(
    csv_file_path: str,
    num_columns: Optional[int] = None,
    preserve_brackets: bool = True,
    merge_excess_into_notes: bool = True,
    notes_merge_start_idx: int = 7,
    fixed_suffix_count: int = 2,
    encoding: str = 'utf-8',
    verbose: bool = True
) -> pd.DataFrame:
    """
    Robustly loads a malformed SME CSV (with unquoted commas in fields) into a clean pandas DataFrame.
    
    Key Features:
    - Handles unquoted commas inside text (especially in Notes, Options)
    - Preserves content inside square brackets [ ] as atomic units
    - Merges excess middle fields into 'Notes' (or first middle column)
    - Pads missing columns with empty strings
    - Validates row lengths
    - Returns fully cleaned DataFrame with correct header
    
    Args:
        csv_file_path (str): Path to the CSV file
        num_columns (int, optional): Expected number of columns. Auto-detected from header if None.
        preserve_brackets (bool): Treat [content] as single unit during split
        merge_excess_into_notes (bool): Merge extra fields in middle into Notes column
        notes_merge_start_idx (int): Column index where "Notes" section begins (0-indexed)
        fixed_suffix_count (int): Number of columns at the end to keep intact (e.g., Evidence?, Confidential?)
        encoding (str): File encoding
        verbose (bool): Print progress and diagnostics
    
    Returns:
        pd.DataFrame: Cleaned dataframe with consistent columns
    """
    
    if verbose:
        print(f"Loading CSV from: {csv_file_path}")
    
    # Read all lines
    try:
        with open(csv_file_path, 'r', encoding=encoding) as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]
    except FileNotFoundError:
        raise FileNotFoundError(f"CSV file not found: {csv_file_path}")
    except Exception as e:
        raise IOError(f"Error reading file: {e}")
    
    if len(lines) == 0:
        raise ValueError("CSV file is empty.")
    if len(lines) == 1:
        raise ValueError("CSV has only header row. No data.")
    
    if verbose:
        print(f"Read {len(lines)} lines (including header)")

    def smart_split_csv_line(line: str, target_cols: int) -> List[str]:
        """
        Split line respecting bracket depth and target column count.
        """
        if not preserve_brackets:
            # Simple comma split fallback
            parts = [p.strip() for p in line.split(',')]
        else:
            # Parse respecting bracket nesting
            parts = []
            current = []
            depth = 0

            for char in line:
                if char == '[':
                    depth += 1
                    current.append(char)
                elif char == ']':
                    depth -= 1
                    current.append(char)
                elif char == ',' and depth == 0:
                    parts.append(''.join(current).strip())
                    current = []
                else:
                    current.append(char)

            if current:
                parts.append(''.join(current).strip())

        # Handle column count mismatches
        if len(parts) == target_cols:
            return parts

        elif len(parts) > target_cols and merge_excess_into_notes:
            # Keep prefix and suffix, merge middle excess into Notes
            prefix = parts[:notes_merge_start_idx]
            suffix = parts[-fixed_suffix_count:] if fixed_suffix_count > 0 else []
            middle = parts[notes_merge_start_idx : -fixed_suffix_count if fixed_suffix_count > 0 else None]

            expected_middle = target_cols - len(prefix) - len(suffix)
            if len(middle) > expected_middle:
                # Merge all excess into the first middle field (Notes)
                notes_field = ', '.join(middle[:-expected_middle + 1])
                rest_middle = middle[-expected_middle + 1:]
                middle = [notes_field] + rest_middle

            return prefix + middle + suffix

        else:
            # Too few columns: pad with empty strings
            while len(parts) < target_cols:
                parts.append('')
            return parts[:target_cols]  # trim if somehow too many after padding logic

    # Parse header
    header_parts = smart_split_csv_line(lines[0], target_cols=100)  # large temp
    if num_columns is None:
        # Try to infer from header — count non-empty meaningful parts
        non_empty = [p for p in header_parts if p and not p.startswith('[')]
        num_columns = len(non_empty)
        if verbose:
            print(f"Auto-detected {num_columns} columns from header.")
    else:
        if verbose:
            print(f"Using user-specified column count: {num_columns}")

    header = smart_split_csv_line(lines[0], target_cols=num_columns)
    if len(header) != num_columns:
        logging.warning(f"Header parsing resulted in {len(header)} columns, expected {num_columns}")

    if verbose:
        print(f"Final header has {len(header)} columns.")

    # Parse data rows
    data = []
    bad_rows = []
    
    for i, line in enumerate(lines[1:], start=2):
        row = smart_split_csv_line(line, target_cols=num_columns)
        if len(row) != num_columns:
            bad_rows.append((i, len(row), row))
            if verbose and len(bad_rows) <= 5:
                print(f"Warning: Row {i} has {len(row)} fields (expected {num_columns})")
        data.append(row)

    # Create DataFrame
    df = pd.DataFrame(data, columns=header)

    # Final validation
    actual_cols = df.shape[1]
    if actual_cols != num_columns:
        if verbose:
            print(f"DataFrame has {actual_cols} columns, expected {num_columns}. Fixing...")
        if actual_cols > num_columns:
            df = df.iloc[:, :num_columns]
            df.columns = header[:num_columns]
        else:
            # Pad columns
            for j in range(actual_cols, num_columns):
                df[header[j] if j < len(header) else f"Unnamed_{j}"] = ''

    # Summary
    if verbose:
        print(f"\n✓ CSV file loaded successfully!")
        print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
        print(f"All rows have {df.shape[1]} columns: {'✓' if (df.shape[1] == num_columns) else '✗'}")

        print("\nColumn Names:")
        print("-" * 80)
        for idx, col in enumerate(df.columns, 1):
            print(f"{idx:2d}. {col}")

        if bad_rows:
            print(f"\nWarning: {len(bad_rows)} rows had column count mismatches (first 5 shown above).")

        print("\nFirst 5 rows preview:")
        print("=" * 120)
        print(df.head(5).to_string(index=False))

    return df
```

```python
import pandas as pd
import re
from typing import List, Optional
import logging

def load_sme_csv_to_dataframe(
    csv_file_path: str,
    num_columns: Optional[int] = None,
    preserve_brackets: bool = True,
    merge_excess_into_notes: bool = True,
    notes_merge_start_idx: int = 7,
    fixed_suffix_count: int = 2,
    encoding: str = 'utf-8',
    verbose: bool = True
) -> pd.DataFrame:
    """
    Robustly loads a malformed SME CSV (unquoted commas, bracketed units) into a clean DataFrame.

    • Works with **any** number of columns – auto-detects the real column count.
    • Preserves `[…]` as atomic units.
    • Optionally merges excess middle fields into the “Notes” column.
    • Pads missing fields with empty strings.

    Args as before – `num_columns` is now truly optional.
    """
    if verbose:
        print(f"Loading CSV from: {csv_file_path}")

    # ------------------------------------------------------------------ #
    # 1. Read raw lines
    # ------------------------------------------------------------------ #
    try:
        with open(csv_file_path, 'r', encoding=encoding) as f:
            lines = [ln.rstrip('\n\r') for ln in f if ln.strip()]
    except FileNotFoundError:
        raise FileNotFoundError(f"CSV file not found: {csv_file_path}")
    except Exception as e:
        raise IOError(f"Error reading file: {e}")

    if not lines:
        raise ValueError("CSV file is empty.")
    if len(lines) == 1:
        raise ValueError("CSV has only a header row – no data.")

    if verbose:
        print(f"Read {len(lines)} lines (including header)")

    # ------------------------------------------------------------------ #
    # 2. Core splitter – respects brackets, works without a target count
    # ------------------------------------------------------------------ #
    def smart_split_csv_line(
        line: str,
        target_cols: Optional[int] = None,
        *,
        preserve_brackets: bool = True
    ) -> List[str]:
        """
        Split a line while honouring nested `[…]`.  
        If *target_cols* is None → return everything that can be split.
        """
        if not preserve_brackets:
            parts = [p.strip() for p in line.split(',')]
        else:
            parts = []
            cur = []
            depth = 0
            for ch in line:
                if ch == '[':
                    depth += 1
                    cur.append(ch)
                elif ch == ']':
                    depth -= 1
                    cur.append(ch)
                elif ch == ',' and depth == 0:
                    parts.append(''.join(cur).strip())
                    cur = []
                else:
                    cur.append(ch)
            if cur:
                parts.append(''.join(cur).strip())

        # -------------------------------------------------------------- #
        # 2a. If we have a target → enforce it (pad / merge)
        # -------------------------------------------------------------- #
        if target_cols is None:
            return parts

        if len(parts) == target_cols:
            return parts

        if len(parts) > target_cols and merge_excess_into_notes:
            # keep prefix + suffix, collapse middle excess into first middle field
            prefix = parts[:notes_merge_start_idx]
            suffix = parts[-fixed_suffix_count:] if fixed_suffix_count else []
            middle = parts[notes_merge_start_idx : -fixed_suffix_count if fixed_suffix_count else None]

            expected_mid = target_cols - len(prefix) - len(suffix)
            if len(middle) > expected_mid:
                excess = middle[: len(middle) - expected_mid + 1]
                notes = ', '.join(excess)
                middle = [notes] + middle[len(middle) - expected_mid + 1 :]
            return prefix + middle + suffix

        # too few → pad
        parts += [''] * (target_cols - len(parts))
        return parts[:target_cols]

    # ------------------------------------------------------------------ #
    # 3. Detect the *real* column count (two-pass)
    # ------------------------------------------------------------------ #
    # First pass: split header without a target → raw header tokens
    raw_header = smart_split_csv_line(lines[0], target_cols=None,
                                      preserve_brackets=preserve_brackets)

    # If the user forced a count → honour it
    if num_columns is not None:
        detected_cols = num_columns
        if verbose:
            print(f"User forced {detected_cols} columns.")
    else:
        # Heuristic: longest *well-formed* row (including header) defines the schema
        candidate_lengths = []
        for ln in lines:
            cand = smart_split_csv_line(ln, target_cols=None,
                                        preserve_brackets=preserve_brackets)
            # ignore rows that are obviously garbage (e.g., a single huge field)
            if len(cand) > 1:
                candidate_lengths.append(len(cand))

        if candidate_lengths:
            detected_cols = max(candidate_lengths)
        else:
            detected_cols = len(raw_header) or 1

        if verbose:
            print(f"Auto-detected {detected_cols} columns (max well-formed row).")

    # ------------------------------------------------------------------ #
    # 4. Build final header (respect detected_cols)
    # ------------------------------------------------------------------ #
    header = smart_split_csv_line(lines[0], target_cols=detected_cols,
                                  preserve_brackets=preserve_brackets)
    # If header is shorter than detected → pad with generic names
    while len(header) < detected_cols:
        header.append(f'Unnamed_{len(header)}')

    if verbose:
        print(f"Final header length: {len(header)}")

    # ------------------------------------------------------------------ #
    # 5. Parse data rows
    # ------------------------------------------------------------------ #
    data_rows = []
    bad_rows = []

    for idx, raw_line in enumerate(lines[1:], start=2):
        row = smart_split_csv_line(raw_line, target_cols=detected_cols,
                                   preserve_brackets=preserve_brackets)
        if len(row) != detected_cols:
            bad_rows.append((idx, len(row), row[:5]))  # keep a tiny preview
            if verbose and len(bad_rows) <= 5:
                print(f"Row {idx}: {len(row)} fields (expected {detected_cols})")
        data_rows.append(row)

    # ------------------------------------------------------------------ #
    # 6. Build DataFrame
    # ------------------------------------------------------------------ #
    df = pd.DataFrame(data_rows, columns=header)

    # ------------------------------------------------------------------ #
    # 7. Final sanity check & summary
    # ------------------------------------------------------------------ #
    if df.shape[1] != detected_cols:
        if verbose:
            print(f"DataFrame column mismatch ({df.shape[1]} vs {detected_cols}). Fixing...")
        if df.shape[1] > detected_cols:
            df = df.iloc[:, :detected_cols]
            df.columns = header[:detected_cols]
        else:
            for j in range(df.shape[1], detected_cols):
                df[header[j]] = ''

    if verbose:
        print("\nCSV loaded successfully!")
        print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
        print("\nColumn names:")
        print("-" * 80)
        for i, col in enumerate(df.columns, 1):
            print(f"{i:2d}. {col}")

        if bad_rows:
            print(f"\nWarning: {len(bad_rows)} rows had wrong column counts (first 5 shown above).")

        print("\nFirst 5 rows preview:")
        print("=" * 120)
        print(df.head(5).to_string(index=False))

    return df
```
